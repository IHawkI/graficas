```python
def remove_duplicates(cell):
    if isinstance(cell, str):
        items = cell.split(', ')
        unique_items = list(dict.fromkeys(items))
        return ', '.join(unique_items)
    return cell
df3= df['Add-ons Purchased'].apply(remove_duplicates)


df_limpio=df3
df4=df_limpio.to_csv('df_limpio.csv', index=False)
df4
```
